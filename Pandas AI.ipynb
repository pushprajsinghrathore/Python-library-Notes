# What is Pandas merge Function ?
Pandas merge connects columns or indexes in
DataFrame based on one or more keys.




```python
df=pd.read_csv("Agriculture data.csv").head()
df
```

```python
df1=pd.DataFrame({'Id':[1,2,3,4],'Class':[9,5,8,6]})
df1
```

```python
df2=pd.DataFrame({'Id':[1,2,3,4],'Name':['A',"b","c","D"]})
df2
```

```python
pd.merge(df1,df2)
```

```python
pd.merge(df1,df2,on="Id")
```

```python
# change the postion of columns
pd.merge(df2,df1,on="Id")
```

```python
pd.merge(df2,df1,on="Id",how="left")
```

```python
pd.merge(df2,df1,on="Id",how="right")
```

```python
pd.merge(df2,df1,on="Id",how="outer")
```

```python
pd.merge(df1, df2, on = 'Id', how = 'outer', indicator = True)
```

```python
df2=pd.DataFrame({"Id":[5,8,9,7],'Name':["A","B","c","D"]})
df2
```

```python
pd.merge(df1,df2,left_index=True,right_index=True)
```

# What is Pandas Join() Method ?
DataFrame join is a convenient method for combining the
columns of two potentially differently-indexed.

```python
df1=pd.DataFrame({"A":[1,2,3],"B":[10,20,30]},index=['a','b','c'])
df2=pd.DataFrame({"C":[1,2,3],"D":[20,30,40]},index=['a','b','c'])
display(df1,df2)
```

```python
df1.join(df2)
```

```python
df1.join(df2)
```
